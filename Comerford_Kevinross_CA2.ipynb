# Environment Setup
To ensure reproducibility and avoid dependency conflicts, all work for this project was carried out in a dedicated Conda environment.

<br>conda create -n ca2_env python=3.10
<br>conda activate ca2_env

All required libraries were then installed using a requirements file:
<br>pip install -r requirements.txt

To enable the environment within Jupyter Notebook, a custom kernel was registered:
<br>python -m ipykernel install --user --name ca2_env --display-name "Python (ca2_env)"

<br>python -m spacy download en_core_web_sm<br>
